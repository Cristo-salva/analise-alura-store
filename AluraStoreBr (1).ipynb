### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

# 4. Produtos Mais e Menos Vendidos

```python
mais_vendidos_1 = loja['Produto'].value_counts().index[0]
qtde_mais_vendidos_1 = int(loja['Produto'].value_counts().values[0])
menos_vendidos_1 = loja['Produto'].value_counts().index[-1]
qtde_menos_vendidos_1 = int(loja['Produto'].value_counts().values[-1])

mais_vendidos_2 = loja2['Produto'].value_counts().index[0]
qtde_mais_vendidos_2 = int(loja2['Produto'].value_counts().values[0])
menos_vendidos_2 = loja2['Produto'].value_counts().index[-1]
qtde_menos_vendidos_2 = int(loja2['Produto'].value_counts().values[-1])

mais_vendidos_3 = loja3['Produto'].value_counts().index[0]
qtde_mais_vendidos_3 = int(loja3['Produto'].value_counts().values[0])
menos_vendidos_3 = loja3['Produto'].value_counts().index[-1]
qtde_menos_vendidos_3 = int(loja3['Produto'].value_counts().values[-1])

mais_vendidos_4 = loja4['Produto'].value_counts().index[0]
qtde_mais_vendidos_4 = int(loja4['Produto'].value_counts().values[0])
menos_vendidos_4 = loja4['Produto'].value_counts().index[-1]
qtde_menos_vendidos_4 = int(loja4['Produto'].value_counts().values[-1])
```

```python
import matplotlib.pyplot as plt
analise_vendas = {
    'Loja 1': {
        'Mais vendidos': {mais_vendidos_1: qtde_mais_vendidos_1},
        'Menos vendidos': {menos_vendidos_1: qtde_menos_vendidos_1}
    },
    'Loja 2': {
        'Mais vendidos': {mais_vendidos_2: qtde_mais_vendidos_2},
        'Menos vendidos': {menos_vendidos_2: qtde_menos_vendidos_2}
    },
    'Loja 3': {
        'Mais vendidos': {mais_vendidos_3: qtde_mais_vendidos_3},
        'Menos vendidos': {menos_vendidos_3: qtde_menos_vendidos_3}
    },
    'Loja 4': {
        'Mais vendidos': {mais_vendidos_4: qtde_mais_vendidos_4},
        'Menos vendidos': {menos_vendidos_4: qtde_menos_vendidos_4}
    }
}
```

```python
produtos_mais_vendidos = {}
for loj, dados in analise_vendas.items():
  for produto, quantidade in dados['Mais vendidos'].items():
    produtos_mais_vendidos[loj + '-' + produto] = quantidade
```

```python
produtos_menos_vendidos = {}
for loj, dados in analise_vendas.items():
  for produto, quantidade in dados['Menos vendidos'].items():
    produtos_menos_vendidos[loj + '-' + produto] = quantidade
```

```python
nomes_produtos_mais_vendidos = list(produtos_mais_vendidos.keys())
quantidades_mais_vendidos = list(produtos_mais_vendidos.values())
```

```python
# Coletar os dados para o gráfico agrupado
lojas_ = list(analise_vendas.keys())
quantidades_mais_vendidos = []
quantidades_menos_vendidos = []
nomes_mais_vendidos = []
nomes_menos_vendidos = []
```

```python
for loj in lojas_:
    # Para cada loja, extrair o produto e a quantidade do mais vendido
    mais_vendido_produto, mais_vendido_qtde = list(analise_vendas[loj]['Mais vendidos'].items())[0]
    quantidades_mais_vendidos.append(mais_vendido_qtde)
    nomes_mais_vendidos.append(mais_vendido_produto)

    # Para cada loja, extrair o produto e a quantidade do menos vendido
    menos_vendido_produto, menos_vendido_qtde = list(analise_vendas[loj]['Menos vendidos'].items())[0]
    quantidades_menos_vendidos.append(menos_vendido_qtde)
    nomes_menos_vendidos.append(menos_vendido_produto)
```

```python
# Célula de Código: Gráfico de Produtos Mais/Menos Vendidos (já existente)
# ... (seu código do gráfico de barras agrupadas)
import numpy as np
largura_barra = 0.35
indice = np.arange(len(lojas_))

plt.figure(figsize=(12, 8))

barras_mais_vendidos = plt.bar(indice - largura_barra/2, quantidades_mais_vendidos, largura_barra, label='Mais Vendido', color='skyblue')
barras_menos_vendidos = plt.bar(indice + largura_barra/2, quantidades_menos_vendidos, largura_barra, label='Menos Vendido', color='lightcoral')

def add_labels(bars, names):
    for bar, name in zip(bars, names):
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, yval + 1, f'{yval} ({name})', ha='center', va='bottom', fontsize=8)

add_labels(barras_mais_vendidos, nomes_mais_vendidos)
add_labels(barras_menos_vendidos, nomes_menos_vendidos)

plt.xlabel('Lojas')
plt.ylabel('Quantidade Vendida')
plt.title('Comparativo de Produtos Mais e Menos Vendidos por Loja')
plt.xticks(indice, lojas_)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show() # Adicione plt.show()
```
